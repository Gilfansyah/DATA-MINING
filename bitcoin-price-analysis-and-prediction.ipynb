```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
file_path = '/kaggle/input/bitcoin-btc-usd-stock-dataset/BTC-USD_stock_data.csv'
df = pd.read_csv(file_path)
df['auto_pct'] = df['Adj Close'].pct_change() * 100
display(df)
```

```python
min_date = min(df['Date'])
max_date = max(df['Date'])

print(f'Min Date: {min_date}')
print(f'Max Date: {max_date}')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['High'], label='High', color='blue')
plt.plot(df['Date'], df['Low'], label='Low', color='orange')
plt.title('High and Low Values of BTC-USD Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks([]) 
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['Date'], df['auto_pct'], color='blue')
plt.xlabel('Date')
plt.ylabel('Percentage Change (%)')
plt.title('Daily Percentage Change in BTC-USD Price')
plt.xticks([])  # Rotate date labels for better readability
plt.tight_layout()

plt.show()
```
